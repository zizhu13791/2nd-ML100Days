# 作業目標:

    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
# 演算法執行並更新
def Gradient_Descent(learning_rate):
    cur_x = 3 # 演算法初始點 x=3
    precision = 0.000001 # 何時停止
    max_iters = 10000 # 最大迭代次數
    df = lambda x: 2*(x+5) # 函數微分

    iters = 0 # 迭代計數器
    previous_step_size = 1
    iters_history = [iters]
    x_history = [cur_x]
    
    while previous_step_size > precision and iters < max_iters:
        prev_x = cur_x
        cur_x = cur_x - learning_rate*df(prev_x)
        previous_step_size = abs(cur_x - prev_x)
        iters = iters+1
    
        print('Iteration',iters,'\nx = ',cur_x)
    
        iters_history.append(iters)
        x_history.append(cur_x)
    
    print('\nTotally iteations：',iters)
    print('The local minimum occurs at',cur_x)
    
    return iters_history, x_history
```

```python
# 畫出圖形
import matplotlib.pyplot as plt

%matplotlib inline

def show_plt(iters_history, x_history, color):
    plt.plot(iters_history, x_history, 'o-', ms=3, lw=1.5, color=color)
    plt.xlabel('iters', fontsize=16)
    plt.ylabel('x value', fontsize=16)
    plt.show()
```

### Learning Rate = 0.1

```python
lr = [0.1, 0.001, 0.0001] # 學習率

iters_history, x_history = Gradient_Descent(lr[0])
```

```python
show_plt(iters_history, x_history, 'black')
```

### Learning Rate = 0.001

```python
iters_history, x_history = Gradient_Descent(lr[1])
show_plt(iters_history, x_history, 'red')
```

### Learning Rate = 0.0001

```python
iters_history, x_history = Gradient_Descent(lr[2])
show_plt(iters_history, x_history, 'blue')
```
