### Rectified Linear Unit- Relu 
    f(x) = max(0,x)


### [Numpy 中 max 與 maximum的差異](https://blog.csdn.net/CSDN5529/article/details/79038544)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    return np.maximum(0,x)
    
x = plt.linspace(-10, 10, 100)
plt.title('ReLU Function')
plt.grid()
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.plot(x, ReLU(x), 'r')
plt.show()
```

```python
def dReLU(x):
    return (x>0) # 這裡的 return 是邏輯 True=1 Flase=0

x = plt.linspace(-10, 10, 100)
plt.title('dReLU Function')
plt.grid()
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.plot(x, dReLU(x), 'r')
plt.show()
```
