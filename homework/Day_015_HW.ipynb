```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline

# 忽略警告訊息
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. 請用 numpy 建立一個 10 x 10, 數值分布自 0.0 ~ 1.0 的矩陣並繪製 Heatmap

"""
Your Code Here
"""
matrix = np.random.random(size = (10,10))

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
#heatmap = 
sns.heatmap(matrix , cmap = 'YlGnBu' , annot = True)
plt.title('Matrix Heatmap')
plt.show()
```

```python
# 2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(nrow,ncol) # 標準分布

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
上半部為 scatter, 對角線為 hist, 下半部為 density
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# 3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(size=(nrow,ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
上半部為 scatter, 對角線為 hist, 下半部為 density
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
