```python
# 引用函式庫
import pandas as pd
import numpy as np
import copy
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 因為擬合(fit)和編碼(transform)需要分開，因此不使用 .get_dummy，而使用 sklearn 的 OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
```

```python
# 指定路徑與檔名
data_path = './data/Part02/'
df = pd.read_csv(data_path + 'titanic_train.csv')

# 存取目標
train_Y = df['Survived']
df = df.drop(['PassengerId' , 'Survived'] , axis=1)

df.head(7)
```

```python
# 需要加入類別型(object)與數值型欄位，所以使用最簡版的特徵工程
LEncoder = LabelEncoder()
MMEncoder = MinMaxScaler()

# 依序每個欄位檢查，有缺值就填補缺值，若是類別型欄位則使用標籤編碼
for col in df.columns:
    df[col] = df[col].fillna(-1)
    
    if df[col].dtype == 'object':
        df[col] = LEncoder.fit_transform( list(df[col].values) )
    
    df[col] = MMEncoder.fit_transform( df[col].values.reshape(-1,1) )
    

df.head(7)
```

```python
train_X = df.values


# 因為訓練邏輯迴歸也需要資料，因此將訓練集切分為三部分 train/val/test，使用 test 驗證而非 k-fold 
# train 用來訓練隨機森林樹，val 用來訓練邏輯斯迴歸，test 驗證效果
train_X, test_X, train_Y, test_Y = train_test_split(train_X, train_Y, test_size=0.5)
train_X, val_X, train_Y, val_Y = train_test_split(train_X, train_Y , test_size=0.5)
```

# 作業1
* 請對照範例，完成隨機森林的鐵達尼生存率預測，以及對應的葉編碼+邏輯斯迴歸

```python
# 隨機森林擬合後, 再將葉編碼 (*.apply) 結果做獨熱 / 邏輯斯迴歸
rf = RandomForestClassifier(n_estimators=20, min_samples_split=10, min_samples_leaf=5, 
                            max_features=4, max_depth=3, bootstrap=True)
onehot = OneHotEncoder()
lr = LogisticRegression(solver='lbfgs', max_iter=1000)

"""
Your Code Here (Hint : 隨機森林的葉編碼(.apply)不需要加上[:, :, 0], 直接用rf.apply()調用即可, 本作業其餘寫法相同)
"""
rf.fit(train_X, train_Y)
onehot.fit(rf.apply(train_X))
lr.fit(onehot.transform(rf.apply(val_X)), val_Y)
```

```python
# 將隨機森林+葉編碼+邏輯斯迴歸結果輸出
"""
Your Code Here 
"""
pred_rf_lr = lr.predict_proba(onehot.transform(rf.apply(test_X)))[:, 1]
fpr_rf_lr, tpr_rf_lr, _ = roc_curve(test_Y, pred_rf_lr)
```

```python
# 將隨機森林結果輸出
"""
Your Code Here 
"""
pred_rf = rf.predict_proba(test_X)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(test_Y, pred_rf)
```

# 作業2
* 上述的結果，葉編碼是否有提高預測的正確性呢?

```python
import matplotlib.pyplot as plt
# 將結果繪圖
"""
Your Code Here 
"""
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_rf, tpr_rf, label='RF')
plt.plot(fpr_rf_lr, tpr_rf_lr, label='RF + LR')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
