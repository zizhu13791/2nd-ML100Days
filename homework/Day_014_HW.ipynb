```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # 另一種繪圖套件

# 在 jupyter 中顯示 matplotlib 繪圖
%matplotlib inline 
plt.style.use('ggplot')

# 忽略警告訊息
import warnings
warnings.filterwarnings('ignore')
```

```python
# 設定 data path
dir_data = './data/home_credit_default_risk/'

# 指定檔名
f_app = os.path.join(dir_data , 'application_train.csv')
print('Path of read in data：%s' %(f_app))

# pandas 讀取 csv 檔
app_train = pd.read_csv(f_app)

app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
house_data = app_train[ ['HOUSETYPE_MODE','AMT_INCOME_TOTAL'] ] # subset
unique_house_type = house_data['HOUSETYPE_MODE'].unique()


print(len(unique_house_type) ,'\n',unique_house_type)
house_data.head()
```

```python
nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[ house_data['HOUSETYPE_MODE'] == unique_house_type[i] , 'AMT_INCOME_TOTAL' ].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```

```python
house_data['AMT_CREDIT'] = app_train['AMT_CREDIT']


plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[ house_data['HOUSETYPE_MODE'] == unique_house_type[i] , 'AMT_CREDIT' ].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show()   
```
