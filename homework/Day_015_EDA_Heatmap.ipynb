```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # 另一種繪圖套件
plt.style.use('ggplot')

%matplotlib inline

# 忽略警告訊息
import warnings
warnings.filterwarnings('ignore')
```

```python
# 設定 data path
dir_data = './data/home_credit_default_risk/'

# 指定檔案名稱
f_app = os.path.join(dir_data , 'application_train.csv')
print('Path of read in data：%s' %(f_app))

# pandas 讀取 csv
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 對出生天數取絕對值
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])

# 取出 EXT_SOURCE 的幾項變數並檢驗其與 TARGET、DAYS_BIRTH 的相關性
ext_data = app_train[ ['TARGET' , 'EXT_SOURCE_1' , 'EXT_SOURCE_2' , 'EXT_SOURCE_3' , 'DAYS_BIRTH'] ]
ext_data_corrs = ext_data.corr()

ext_data_corrs
```

```python
plt.figure( figsize = (8,6) )

# 繪製相關係數的 Heatmap
#sns.heatmap( ext_data_corrs , cmap=plt.cm.RdYlBu_r , vmin=-0.25 , annot = True, vmax = 0.6 )
sns.heatmap( ext_data_corrs , cmap = 'YlGnBu' , annot = True)
plt.title('Correlation Heatmap')
```

**[enumerate( )的用法](http://www.runoob.com/python/python-func-enumerate.html)**


tight_layout()可以用來調整圖形間距：  
matplotlib.pyplot.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)

例如： plt.tight_layout()  #自动调整subplot间的参数

```python
# 進一步檢查三項 EXT_SOURCE 相關變數在 TARGET 上的分布
plt.figure( figsize = (24,8) )

# 依照不同的 EXT_SOURCE 逐項繪製 KDE 圖形
for i , source in enumerate( ['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3'] ):
    # 做 subset，放在指定的位置
    plt.subplot(1 , 3 , i+1)
    
    # 繪製 KDE 圖形
    sns.kdeplot( app_train.loc[app_train['TARGET'] == 0 , source] , label = 'target = 0' )
    sns.kdeplot( app_train.loc[app_train['TARGET'] == 1 , source] , label = 'target = 1' )
    
    # 對圖形加上標籤
    plt.title('Distribution of %s by Target Value' %source)
    plt.xlabel('%s' %source)
    plt.ylabel('Density')
    
plt.tight_layout(h_pad = 2.5)
```

```python
# 將 subset ext_data 資料複製到 plot_data
plot_data = ext_data.copy()
print(plot_data.head())
# 將出生天數轉化成年齡
plot_data['YEARS_BIRTH'] = plot_data['DAYS_BIRTH']/365

# 除去日齡紀錄
plot_data.drop(['DAYS_BIRTH'] , axis=1 , inplace=True)

plot_data.head()
```

## **[annotate( ) ](https://www.itread01.com/content/1544608110.html)**  
用於在圖形上給資料新增文字註解，而且支援帶箭頭的劃線工具，方便我們在合適的位置新增描述資訊。

```python
# 定義函數：計算兩兩 column 之間的相關係數
def corr_func(x , y , **kwargs):
    r = np.corrcoef(x , y)[0][1]
    
    ax = plt.gca()  # 獲得當前座標軸的訊息
    ax.annotate('r = {:.2f}'.format(r) ,
                xy = (.2 , .8) ,
                xycords = ax.transAxes ,
                size = 20)
```

```python
N_sample = 100000

# 將 NaN 數值刪除，並限制資料上限為 100000，因為要畫點圖，如果點太多會耗時間
plot_data = plot_data.dropna().sample(n = N_sample)
```

```python
# 建立 pairgrid 物件
grid = sns.PairGrid( data = plot_data , size = 3 , diag_sharey = False ,
                     hue = 'TARGET' ,
                     vars = [x for x in list(plot_data.columns) if x != 'TARGET'])

# 上半部為 scatter
grid.map_upper( plt.scatter , alpha = 0.2 )

# 對角線畫 histogram
grid.map_diag( sns.kdeplot )

# 下半部畫 density plot
grid.map_lower( sns.kdeplot , cmap = plt.cm.OrRd_r )


plt.suptitle('Ext Source and Age Features Pairs Plot' , size = 32 , y = 1.05)
plt.show()
```

```python
# 嘗試以不同的 size 畫看看，觀察差異
N_sample = 1000

plot_data = plot_data.dropna().sample(n = N_sample)

# 建立 pairgrid 物件
grid = sns.PairGrid(data = plot_data , size = 3 , diag_sharey = False ,
                    hue = 'TARGET' ,
                    vars = [x for x in list(plot_data.columns) if x != 'TARGET'])


# 上半部為 scatter
grid.map_upper( plt.scatter , alpha = 0.2 )

# 對角線畫 histogram
grid.map_diag( sns.kdeplot )

# 下半部畫 density plot
grid.map_lower( sns.kdeplot , cmap = plt.cm.OrRd_r )


plt.suptitle('Ext Source and Age Features Pairs Plot' , size = 32 , y = 1.05)
plt.show()
```
