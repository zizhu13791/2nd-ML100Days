### 練習時間 - 資料擷取

**取得指定欄位及指定列數的資料**

````python
app_train[['SK_ID_CURR','FLAG_OWN_REALTY','CNT_CHILDREN','ORGANIZATION_TYPE']][50:60]
````

**取得`所有欄位`中的第100~110筆的資料內容**

````python
app_train.iloc[:][100:111]
# 寫法同 app_train.iloc[100:111 , :]
````

**NaN(Not a Number)即代表是遺失值，最常見的兩種方式處理就是：**  
* `DataFrame.dropna( )`： 刪除遺失值
* `DataFrame.fillna( )`： 補值(通常會補平均值)

````python
# 在上一個練習中第 108 筆資料中含有 NaN
# 使用 dropna 將欄位 AMT_REQ_CREDIT_BUREAU_DAY 含有 NaN 的數據列刪除
app_train.iloc[:][100:111].dropna(axis='index',how='any',subset=['AMT_REQ_CREDIT_BUREAU_DAY'])
````

````python
# 在上一個練習中第 108 筆資料中含有 NaN
# 使用 fillna 將欄位 AMT_REQ_CREDIT_BUREAU_DAY 含有 NaN 的數據補值
app_train.iloc[:][100:111].fillna(np.mean(app_train[['AMT_REQ_CREDIT_BUREAU_DAY']]))
````

````python
# 在上一個練習中第 108 筆資料中含有 NaN
# 使用 fillna 將欄位含有 NaN 的數據補 0 值
app_train.iloc[:][100:111].fillna(0)
````
